```python
#import libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

#<center>**Read Data**</center>

```python
dfBTC=pd.read_csv("BTC-EUR.csv")
dfETH=pd.read_csv("ETH-EUR.csv")
```

```python
dfBTC.head(3)
```

```python
dfETH.head(3)
```

```python
print(f"shape of dataBTC: {dfBTC.shape}\nshape of dataETH: {dfETH.shape} ")
```

```python
#All Type Of each features in data
dfBTC.info()
```

```python
#Change type of Date to dateTime
dfBTC.Date= pd.to_datetime(dfBTC.Date)
dfETH.Date= pd.to_datetime(dfETH.Date)
```

#<center>**Data Analyse**</center>

```python
dfBTC.describe()
```

```python
dfETH.describe()
```

## ***Data Exploration***

```python
sns.displot(data=dfBTC,x="Open",kde=True)
plt.title("Open Distribution in dfBTC")
plt.show()
```

```python
sns.displot(data=dfETH,x="Open",kde=True)
plt.title("Open Distribution in dfETH")
plt.show()
```

```python
sns.displot(data=dfBTC,x="Close",kde=True)
plt.title("Close Distribution in dfBTC")
plt.show()
```

```python
sns.displot(data=dfETH,x="Close",kde=True)
plt.title("Close Distribution in dfETH")
plt.show()
```

## ***Features Correlation***

```python
sns.heatmap(dfBTC.corr(),annot=True)
plt.show()
```

```python
sns.heatmap(dfETH.corr(),annot=True)
plt.show()
```

```python
#TimeSeries Analyse
px.line(dfBTC,x="Date",y="Volume",title="Distribution of Volume of BTC @Time ")
```

```python
#TimeSeries Analyse
px.line(dfETH,x="Date",y="Volume",title="Distribution of Volume of ETH @Time",color_discrete_sequence=["brown"])
```

```python
#btcVseth=pd.DataFrame({'date':dfETH.Date,'BTC':dfBTC.Close,'ETH':dfETH.Close})
btcVseth=pd.merge(dfBTC[["Date","Close"]],dfETH[["Date","Close"]],on="Date",how="right")
btcVseth=btcVseth.fillna(0)
btcVseth
```

```python
px.line(btcVseth,x="Date",y=["Close_x","Close_y"],title="Evelution of close price each cryptomoney @Time",labels={"value":"close price"})
```
